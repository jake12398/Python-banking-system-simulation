```python
import os
import json
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from IPython.display import HTML, display

interest = 3.75


# =========================================================
# FUNCTIONS (ONLY TO REMOVE REPETITION)
# =========================================================

def apply_interest(balance, balance_hist):
    last_timestamp_str = max(balance_hist.keys())
    last_timestamp = datetime.strptime(last_timestamp_str, "%Y-%m-%d %H:%M:%S")

    days_elapsed = (datetime.now() - last_timestamp).total_seconds() / (24 * 60 * 60)
    return balance * (1 + interest / 365) ** days_elapsed


def update_balance(balance_hist, balance):
    date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    balance_hist[date] = balance


def add_transaction(transactions, t_type, amount):
    date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    transactions[date] = {"type": t_type, "amount": amount}


# =========================================================
# LOAD USERS
# =========================================================

running = True

if os.path.exists("users.json"):
    with open("users.json", "r") as file:
        usernames = json.load(file)
else:
    usernames = {}


# =========================================================
# MAIN LOOP
# =========================================================

while running:

    running2 = True

    while running2:

        print("\nPlease select an account to view")

        for number, value in enumerate(usernames.keys(), start=1):
            print(f"{number}. {value.capitalize()}")

        print("If you wish to create an account, type *")
        print("If you wish to exit, type @")

        Username_input = input("Select an account: ").strip()

        matched_user = None

        for key in usernames:
            if key.lower() == Username_input.lower():
                matched_user = key
                break

        if matched_user:
            pin = usernames[matched_user]
            break

        elif Username_input == "*":
            name_input = input("Please enter a name for this account: ").strip()
            name = name_input.capitalize()

            if name in usernames:
                print("Account already exists.")
                continue

            while True:
                password = input("Now please enter a 4 digit pin: ").strip()
                if len(password) == 4 and password.isdigit():
                    break
                else:
                    print("PIN must be exactly 4 digits.")

            usernames[name] = password
            print(f"Account '{name}' created successfully!")
            continue

        elif Username_input == "@":
            print("Goodbye")
            running = False
            running2 = False
            break

        else:
            print("Account does not exist. Please try again.")

    if not running:
        break

    guesses = 3
    is_running = True

    transactions_file = f"{matched_user}'s transaction history"
    balance_file = f"{matched_user}'s balance history"
    standingorder_file = f"{matched_user} Standing orders"

    # -------------------------
    # LOAD DATA
    # -------------------------

    transactions = {}
    standingorder = {}

    if os.path.exists(transactions_file):
        with open(transactions_file, "r") as f:
            transactions = json.load(f)

    if os.path.exists(standingorder_file):
        with open(standingorder_file, "r") as f:
            standingorder = json.load(f)

    if os.path.exists(balance_file) and os.path.getsize(balance_file) > 0:
        with open(balance_file, "r") as f:
            balance_hist = json.load(f)

        balance = balance_hist[max(balance_hist)]
        balance = apply_interest(balance, balance_hist)

    else:
        balance = 1000
        balance_hist = {}
        update_balance(balance_hist, balance)

    # -------------------------
    # PIN CHECK
    # -------------------------

    while guesses > 0:
        pin_input = input(f"Please enter your PIN. You have {guesses} guesses remaining: ")

        if pin_input == pin:
            break

        guesses -= 1
        print("Incorrect PIN")

    if guesses == 0:
        print("You have been locked out.")
        exit()

    # =========================================================
    # MENU LOOP (UNCHANGED)
    # =========================================================

    while is_running:

        print(f"\nYour balance is £{balance:.2f}. Please select one of the following")
        print("1. Check balance")
        print("2. Deposit money")
        print("3. Withdraw money")
        print("4. Pay someone in the UK")
        print("5. View Transaction history")
        print("6. Change PIN")
        print("7. View or create a standing order")
        print("8. Log out")
        print("9. Exit")

        option = input("Please select a number 1-8: ")

        # -------------------------
        # 1 CHECK BALANCE
        # -------------------------
        if option == "1":
            print(f"Your balance is £{balance:.2f}")

        # -------------------------
        # 2 DEPOSIT
        # -------------------------
        elif option == "2":
            deposit = int(input("How much would you like to deposit? "))

            if deposit > 0:
                balance = apply_interest(balance, balance_hist)
                balance += deposit
                update_balance(balance_hist, balance)
                add_transaction(transactions, "deposit", deposit)

        # -------------------------
        # 3 WITHDRAW
        # -------------------------
        elif option == "3":
            withdraw = int(input("How much would you like to withdraw? "))

            if 0 < withdraw <= balance:
                balance = apply_interest(balance, balance_hist)
                balance -= withdraw
                update_balance(balance_hist, balance)
                add_transaction(transactions, "withdraw", withdraw)

        # -------------------------
        # 4 TRANSFER
        # -------------------------
        elif option == "4":
            print("Please select an account to send money to:")

            for key in usernames:
                if key != matched_user:
                    print(key)

            selecteduser = input("Selected account: ")

            if selecteduser in usernames:
                amount = float(input("Amount: "))

                if 0 < amount <= balance:
                    balance = apply_interest(balance, balance_hist)
                    balance -= amount
                    update_balance(balance_hist, balance)
                    add_transaction(transactions, "transfer", amount)

        # -------------------------
        # 5 TRANSACTIONS
        # -------------------------
        elif option == "5":
            print("Transaction History:")
            for ts, t in transactions.items():
                if t["type"] == "deposit":
                    display(HTML(f"<span style='color:green'>+{t['amount']}</span>"))
                else:
                    display(HTML(f"<span style='color:red'>-{t['amount']}</span>"))

        # -------------------------
        # 6 CHANGE PIN
        # -------------------------
        elif option == "6":
            current_pin = input("Enter current PIN: ")

            if current_pin == pin:
                new_pin = input("Enter new PIN: ")

                if len(new_pin) == 4 and new_pin.isdigit():
                    usernames[matched_user] = new_pin
                    pin = new_pin

        # -------------------------
        # 7 STANDING ORDERS
        # -------------------------
        elif option == "7":
            choiceso = input("1 view / 2 create")

            if choiceso == "1":
                if not standingorder:
                    print("No standing orders found.")
                else:
                    for ts, order in standingorder.items():
                        print(f"{order['reference']}: £{order['amount']} every {order['days']} days (created {ts})")

            elif choiceso == "2":
                days = int(input("Days: "))
                amount = float(input("Amount: "))
                reference = input("Reference: ")

                date = datetime.now().strftime("%Y-%m-%d  %H:%M")
                standingorder[date] = {
                    "amount": amount,
                    "days": days,
                    "reference": reference
                }

        # -------------------------
        # 8 LOG OUT
        # -------------------------
        elif option == "8":
            is_running = False

        # -------------------------
        # 9 EXIT (FIXED EXACT BEHAVIOUR)
        # -------------------------
        elif option == "9":
            is_running = False
            running = False

            print(f"Final balance: £{balance:.2f}. Goodbye.")

            # ASK DAYS (YOUR ORIGINAL BEHAVIOUR)
            try:
                vdays = int(input("View your balance over the last how many days? "))
            except ValueError:
                print("Invalid input. Defaulting to 7 days.")
                vdays = 7

            end_date = datetime.now()
            start_date = end_date - timedelta(days=vdays)

            x = [datetime.strptime(ts, "%Y-%m-%d %H:%M:%S") for ts in balance_hist.keys()]
            y = list(balance_hist.values())

            filtered = [(d, b) for d, b in zip(x, y) if d >= start_date]

            if not filtered:
                print("No data in that range.")
            else:
                x_filtered, y_filtered = zip(*filtered)

                plt.figure(figsize=(10, 5))
                plt.plot(x_filtered, y_filtered, marker='o')

                plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
                plt.gcf().autofmt_xdate()

                plt.xlabel("Time")
                plt.ylabel("Balance (£)")
                plt.title("Balance Over Time")
                plt.grid(True)

                plt.ylim(0, max(y_filtered) + 500)
                plt.show()

    # =========================================================
    # SAVE EVERYTHING
    # =========================================================

    with open(transactions_file, "w") as f:
        json.dump(transactions, f)

    with open(balance_file, "w") as f:
        json.dump(balance_hist, f)

    with open(standingorder_file, "w") as f:
        json.dump(standingorder, f)

    with open("users.json", "w") as f:
        json.dump(usernames, f)
```
